```python
import pandas as pd
```

```python
F = pd.read_csv('ES113_Assignment1.csv')
F
```

# **Question 1**

```python
X = F[F['Gender']=='Female']

#Mean of English Marks
ME = X['English'].mean()
#Mean of Mathematics Marks
MM = X['Mathematics'].mean()
#Mean of Science Marks
MS = X['Science'].mean()
#Mean
Mean = (ME+MM+MS)/3
print("Average= " + str(Mean))
```

# **Question 2**

```python
Median = F['Mathematics'].median()
X = F[F['Mathematics']>Median]

Count = 0
for i in X['Name']:
  Count +=1

print("Answer = " + str(Count))
```

# **Question 3**

```python
Mode = F['Science'].mode()
print("Answer = " + str(Mode))
```

# **Question 4**

```python
Mean = F['English'].mean()
Y = F[F['Gender']=='Male']
X = Y[Y['English']<Mean]

count = 0
for i in X['Name']:
  count += 1

print("Answer = " + str(count))
```

# **Question 5**

```python
X = F[F['Age']==15]

SM = X['Mathematics'].sum()
SS = X['Science'].sum()
SE = X['English'].sum()
print("Answer = " + str(SM+SS+SE))
```

# **Question 6 **

```python
X=F[F.Gender=="Female"]
Y=X[(X['Mathematics'].between(85,90)) | (X['Science'].between(85,90)) | (X['English'].between(85,90))]
print("Answer = " + str(len(Y)))
```

# **Question 7**

```python
X = F[F['Name'].str[0]=='A']
MM = X['Mathematics'].mean()
MS = X['Science'].mean()
ME = X['English'].mean()
Mean = (MM+MS+ME)/3
print("Answer = " + str(Mean))
```

# **Question 8 **
